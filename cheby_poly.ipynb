```python
import numpy as np
from matplotlib import pyplot as plt

N = 32769

x = np.linspace(-1, 1, N)
cheby_poly_lut_0 = np.ones(N)
cheby_poly_lut_1 = np.linspace(-1, 1, N)
waveshape = cheby_poly_lut_1

plt.plot(waveshape)
plt.show()
```

```python
for i in range(2, 4+1):
    next_cheby_poly_lut = 2*x*cheby_poly_lut_1-cheby_poly_lut_0
    cheby_poly_lut_0 = cheby_poly_lut_1
    cheby_poly_lut_1 = next_cheby_poly_lut

    plt.plot(next_cheby_poly_lut)
    plt.show()
```
